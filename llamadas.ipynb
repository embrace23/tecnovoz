```python
#Paso los porcentajes a strings para que se pasen correctamente cuando los exporto y poder trabajar con ellos

datos["Porc. en llamadas"] = datos["Porc. en llamadas"].astype(str)
datos["Porc. no disp"] = datos["Porc. no disp"].astype(str)
datos["Porc. disp"] = datos["Porc. disp"].astype(str)
```

```python
#Creo el conjunto de operadores por Expertice
supervisores = ["DOS SANTOS DOMINGUES CAMILA CAROLINA", "GONZALEZ FERNANDO EZEQUIEL", "MARTINO NICOLAS EZEQUIEL", 
                "OROPEZA RODRIGUEZ ESTHEFANIA IVELISSA", "RENGEL PEDRO FELIPE"]
team_leader = ["GANDARA MAXIMO", "LARRIEL GAYOSO LAURA SOLEDAD", "MARTINO IVAN EZEQUIEL", "MATHEUS SABINO FARIA"]
senior = ["COSTA SANTOS SILVA LORRANA", "DUEÑAS RODRIGUEZ ALEXANDER WASHINGTON",  "GARAVENTA ANIBAL DANIEL", "GODOY CARVAJAL ANNA PAOLA", "GONGO RENATO CANDIDO GONCALVES",
          "GONZALEZ MARTIN ESTEBAN", "LOBO SAENZ ELIANGEL", "MARTINEZ LEANDRO GABRIEL", "MARTINEZ RAMIREZ SINAI DANIELA", "PARLAGRECO MATIAS", "PISANI MARIANO NICOLAS", 
          "QUIROGA DANIELA MARIBEL"]

# Definir la función para asignar el valor de "expertice" en función de "Operador"
def asignar_expertice(row):
    operador = row['Operador']
    if operador in team_leader:
        return 'Team Leader'
    elif operador in senior:
        return 'Senior'
    elif operador in supervisores:
        return 'Sup'
    else:
        return 'Junior'

# Aplico la función a la columna Expertice
datos['Expertice'] = datos.apply(asignar_expertice, axis=1)
```

```python
#Creo el conjunto de operadores por Turno
maniana = ["CAVASSIM DANILO", "CESAR PRADO", "CHARRIER LUIZ RAFAEL", "COSTA SANTOS SILVA LORRANA", "DA COSTA PEREIRA MAYARA", "DA FONSECA MIGUEL", "DA MATA ARGUELHO GUILHERME HENRIQUE", 
           "DOS SANTOS DOMINGUES CAMILA CAROLINA", "DUEÑAS RODRIGUEZ ALEXANDER WASHINGTON", "ESTIVILL SOSENKO ISMAEL", "FERNANDES TOME LUANA VANESSA", "GANDARA FRANCO", 
           "GODOY CARVAJAL ANNA PAOLA", "HOBERT DELFINA", "LUCAS ORELLANA", "MARTINEZ RAMIREZ SINAI DANIELA", "MARTINO IVAN EZEQUIEL", "MARTINS DA SILVA MARIA EDUARDA", "ONORATI BELLI AGOSTINA AILEN", 
           "PISANI MARIANO NICOLAS", "PORTO MACHADO ELICA", "SOFIA TEIXEIRA PORTELLA"]
noche = ["BRUNO LOPES BATISTA", "CAMILA DE ASSIS PAIVA", "CASTELLANO MATIAS NAHUEL", "EVANGELISTA MARCIO JOSE", "IBARRA GARCIA CARLOS ANDRES", "LETICIA LORETA ROQUE STETTER", "ROMINA TEJERINA",
         "ROSILLO HERRERA GUSTAVO", "TOSTA ANDRYEL LUIZ", "URIBE REBECCA"]
dia_sdf = ["CORDEIRO TRINDADE IUCARA", "CUNHA DE ALMEIDA FRANCISCA ELISANGELA", "DA SILVA SANTOS EMILY DAYANE", "FERREIRA DE SANTANA CARVALHO GABRIELLE", "FLANAGAN PAOLA ANDREA", 
           "GAMEZ FERNANDEZ ANGEL GABRIEL", "GARAVENTA ANIBAL DANIEL", "GONCALVES LIMA ADELINA KATIA", "LOBO SAENZ ELIANGEL", "LOPEZ MICAELA SOLEDAD", "MARQUES PEREIRA DARA", 
           "MATHEUS SABINO FARIA", "MEDINA GISELA SABRINA", "QUIROGA DANIELA MARIBEL", "SEMONI GUIDO"]
noche_sdf = ["RAMOS DOS SANTOS AKLA KEZIA", "ROMERSTEIN MARIA FLORENCIA", "RUIZ NATALIA", "SILVA DE ATAIDE ROBERTO"]

# Definir la función para asignar el valor de "Turno" en función de "Operador"
def asignar_turno(row):
    operador = row['Operador']
    if operador in maniana:
        return 'Mañana'
    elif operador in noche:
        return 'Noche'
    elif operador in dia_sdf:
        return 'Dia (SDF)'
    elif operador in noche_sdf:
        return 'Noche (SDF)'
    else:
        return 'Tarde'

#Aplico la función a la columna Turno
datos['Turno'] = datos.apply(asignar_turno, axis=1)
```

```python
#Creo el conjunto de operadores por Equipo
riesgo_propio = ["FERNANDES TOME LUANA VANESSA", "FLANAGAN PAOLA ANDREA", "GARAVENTA ANIBAL DANIEL", "GODOY CARVAJAL ANNA PAOLA"]
riesgo_tpa = ["CASTELLANO MATIAS NAHUEL", "DUEÑAS RODRIGUEZ ALEXANDER WASHINGTON", "FERREIRA ADRIEL", "GAMEZ FERNANDEZ ANGEL GABRIEL", "GANDARA FRANCO", "GONZALEZ MARTIN ESTEBAN", 
              "GUERRERO CHACON MAYLU DANIELA" "IBARRA GARCIA CARLOS ANDRES", "LARRIEL GAYOSO LAURA SOLEDAD", "MARTINEZ BRACHO ANGEL GABRIEL", "MARTINEZ LEANDRO GABRIEL", 
              "MEDINA GISELA SABRINA", "PISANI MARIANO NICOLAS"]
riesgo_tpa_br = ["PORTO MACHADO ELICA"]
riesgo_tpa_inter = ["GANDARA MAXIMO", "LOBO SAENZ ELIANGEL", "LOPEZ MICAELA SOLEDAD", "MARTINEZ RAMIREZ SINAI DANIELA", "QUIROGA DANIELA MARIBEL", "ROMERSTEIN MARIA FLORENCIA",
                    "SEMONI GUIDO", "URIBE REBECCA"]
tpa = ["ESTIVILL SOSENKO ISMAEL", "FACUNDO JOAQUIN BORGNIS MARTINEZ", "FONTAN JUAN PABLO ESTEBAN", "HOBERT DELFINA", "PARLAGRECO MATIAS", "ROSILLO HERRERA GUSTAVO", "RUIZ NATALIA"]
tpa_inter = ["ONORATI BELLI AGOSTINA AILEN"]

# Definir la función para asignar el valor de "Equipo" en función de "Operador"
def asignar_equipo(row):
    operador = row['Operador']
    if operador in riesgo_propio:
        return 'Riesgo propio'
    elif operador in riesgo_tpa:
        return 'Riesgo-TPA'
    elif operador in riesgo_tpa_br:
        return 'Riesgo-TPA-BR'
    elif operador in riesgo_tpa_inter:
        return 'Riesgo-TPA-Inter'
    elif operador in tpa:
        return 'TPA'
    elif operador in tpa_inter:
        return 'TPA-Inter'
    else:
        return 'Riesgo propio-BR'

#Aplico la función a la columna Turno
datos['Equipo'] = datos.apply(asignar_equipo, axis=1)
```

```python
# Definir las columnas en el orden deseado
columnas_deseadas = ['Operador', 'Expertice', 'Turno', 'Equipo','Porc. en llamadas', 'Entrantes', 'Manuales', 'Salientes', 'Porc. disp', 'No disp (HR)', 'Porc. no disp', 'No respondidas']

# Reordenar las columnas del DataFrame
datos = datos.reindex(columns=columnas_deseadas)
```

#### Trabajo sobre el nombre del archivo a descargar

```python
# Establecer la configuración regional en español para que el mes se registre en español
locale.setlocale(locale.LC_TIME, 'es_ES.UTF-8')

# Obtener la fecha actual
fecha_actual = datetime.now()

# Obtener el número del mes actual
numero_mes_actual = fecha_actual.strftime("%m")

# Obtener el nombre del mes actual
nombre_mes_actual = fecha_actual.strftime("%B").capitalize()

# Obtener el año actual
anio_actual = fecha_actual.year

# Obtener el día del mes actual
dia_actual = fecha_actual.day

# Determinar el tipo de quincena
if dia_actual >= 1 and dia_actual <= 15:
    tipo_quincena = "2da quincena"
else:
    tipo_quincena = "1ra quincena"

# Generar el nombre del archivo con el formato deseado
nombre_archivo = f"{numero_mes_actual}. {nombre_mes_actual} {anio_actual} - {tipo_quincena}.csv"
```

```python
#Chequeo que sea el numero de mes correcto
print("Numero de mes: ", numero_mes_actual)
#Chequeo que sea el mes correcto
print("Mes: ", nombre_mes_actual)
#Chequeo que sea el año actual
print("Año: ", anio_actual)
#Chequeo que sea la quincena correcta
print(tipo_quincena)
#Veo el nombre del archivo
print(nombre_archivo)
```

```python
#Descargo la tabla general
datos.to_csv(nombre_archivo, index=False)

#Descargo la tabla SDF
datos2.to_csv(nombre_sdf, index=False)

#Descargo la tabla semana
datos3.to_csv(nombre_semana, index=False)

#Descargo la tabla sup
datos4.to_csv(nombre_sup, index=False)
```
